```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embedder = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
text = "Once upon a time, there was a little"
embedded_text = embedder.embed_query(text)
embedded_text
```

```python
len(embedded_text)
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("mypdf.pdf")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
```

```python
splitted_docs = text_splitter.split_documents(docs)
splitted_docs[0]
```

```python
from langchain_community.vectorstores import Chroma
```

```python
db = Chroma.from_documents(splitted_docs,embedder)
```

```python
searched_data=db.similarity_search("When we get back the text")
#searched_data = db.similarity_search("When we get back the text, the first phase is to assign the speaker names to the\
#speaker labels")
```
